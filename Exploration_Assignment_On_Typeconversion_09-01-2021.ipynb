```python
# Exploration Assignment:

# Complex -  int,float,boolean,str
# boolean - int,float,complex,str
# str(numeric) - int,float,complex,boolean
# str(non-numeric) - int,float,complex,boolean

# Complex conversion

b = complex(23.3) # assignment of b
print('############# Complex number type conversion ############')

print('Variable b type:',type(b))
try:
    print(int(b)) 
except:
    print("Cannot convert complex number to int")
    
try:
    print(float(b)) 
except:
    print("Cannot convert complex number to float")
    
print(str(b))
print(bool(b))

# Boolean conversion

b = True # reassignment of b
print('############# Boolean type conversion ###########')
print('Variable b type:',type(b))
print(int(b)) 
print(float(b)) 
print(str(b))
print(complex(b))


b = str(12) # reassignment of b
print('############# Str(number) type conversion')
print('Variable b type:',type(b))
print(int(b)) 
print(float(b))
print(bool(b))
print(complex(b))

 

b = str('hello world1') # reassignment of b

print('############# Str(non-number) type conversion')
print('Variable b type:',type(b))
print(type(b))
print(bool(b))
try:
    print(complex(b)) 
except:
    print("Cannot convert str(non-numeric) to complex")

try:
    print(int(b)) 
except:
    print("Cannot convert str(non-numeric) to int")
    
try:
    print(float(b)) 
except:
    print("Cannot convert str(non-numeric) to float")
    
```
